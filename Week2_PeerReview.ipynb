# My Equation

```julia
f(x) = (x + 5) * (x - 3)
```

$$f(x) = (x + 5)(x - 3)$$

```julia
# print heading
println("x    f(x)")

# define empty arrays for x and y axes, which will be populated by the for loop
xaxis = (Int64)[]
yaxis = (Int64)[]

# print range of x from -10 to 8, with f(x)
# also fill out the arrays with x and y values
# the push!() function adds a value to the arrays each time
# thus xaxis = range of x values (-10 to 8)
# yaxis = f(x) corresponding to x values
plotrange = -10:8
for i in plotrange 
    println("$i    $(f(i))")
    push!(xaxis, i)
    push!(yaxis, f(i))
end
```

```julia
# Plot data
using Plots
gr()
plot(xaxis, yaxis, title = "f(x) = (x+5)(x-3)", label = "f(x)", legend = :bottomright, marker = (:circle, 5))
plot!(zero, label = "zero line")
```
