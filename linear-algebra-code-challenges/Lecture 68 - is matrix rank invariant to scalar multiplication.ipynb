```julia
using LinearAlgebra
```

```julia
# create two matrices, 1 full rank and 1 reduced rank
# both rectangular

A = rand(-20:20, 4,7)
```

```julia
B = rand(-20:20, 4,7);
B[1,:] .= 0'
B
```

```julia
rank(B)
```

```julia
scalar = rand(-20:20)
if scalar == 0
    scalar += 1
end
```

```julia
scalar
```

```julia
C = scalar * A
```

```julia
D = scalar * B
```

```julia
rank(A)
```

```julia
rank(B)
```

```julia
x = collect(-20:0.1:20)
```

```julia
y = []
for i in x
    push!(y, rank(A*i))
end
```

```julia
using Plots; gr()
```

```julia
plot(x,y)
```

```julia
w = []
for i in x
    push!(w, rank(B*i))
end
```

```julia
plot(x,w)
```

## Thus with the exception of zero, matrix rank does not vary with scalar
